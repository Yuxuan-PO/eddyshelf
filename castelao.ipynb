# Non-dimensional parameters for eddies

Looking at Renato's dataset. My goal is to do some non-dimensional parameter statistics

1. Compare eddy radius $L$ to deformation radius $L_D$ in region
2. Rossby, Rhines number statistics
3. Radius / Slope width

The deformation radius is obtained from the database of Chelton et al. (1998). I filter for WCRs by checking whether the center crosses the 4000m isobath at least once.

```python
%matplotlib inline
%load_ext secnum
%secnum

import numpy as np
import scipy as sp
import h5py
import netCDF4 

import numexpr as ne
from scipy.io import loadmat
import scipy.interpolate

from datetime import datetime
import time
from IPython import display
import pylab

# matplotlib section
import matplotlib as mpl
import matplotlib.pyplot as plt
# big figures
mpl.rcParams['savefig.dpi'] = 2 * mpl.rcParams['savefig.dpi']
# better color cycling
import brewer2mpl
colors =  brewer2mpl.get_map('Set2', 'qualitative', 8).mpl_colors
mpl.rcParams['axes.color_cycle'] = colors
```

```python
# Read in Castelao datasets
# this file has trajectories folowing each eddy
file2 = 'data/castelao/03eddy_trajectories_7days_gap40km_1992_Leff_mod_MAB_v5.mat'
trajs = loadmat(file2)
# trajs['eals'] is Ls in Chelton et al. (2011) notation. 
# convert to metres
Ls = trajs['eals'][0]*1000
U = trajs['eaV'][0]
lon = trajs['eaX'][0]
lat = trajs['eaY'][0]

# load Chelton et al. (1998) Rossby radius atlas
rossfile = '../datasets/rossbyradiusatlas.mat'
ross = loadmat(rossfile)
# grid form for griddata call later
rossgrid = np.c_[ross['lon'], ross['lat']]
# meshgrid to interpolate on to
rossx,rossy = np.meshgrid(np.arange(rossgrid[:,0].min(),rossgrid[:,0].max()),
                          np.arange(rossgrid[:,1].min(),rossgrid[:,1].max()))
# grid deformation radius onto uniform grid
defrad = sp.interpolate.griddata(rossgrid, ross['L'], (rossx, rossy), method='nearest').squeeze().T

# load ETOPO1 bathymetry
bathyfile = 'data/etopo2_extract.mat'
bathy = loadmat(bathyfile, struct_as_record=False, squeeze_me=True)
topo = bathy['topo'] # pull out the topo structure
```

```python
def find_nearest(a, a0):
    return np.abs(a - a0).argmin()
```

```python
# estimate Ro, Rh, deformation radius and reduce to one value per eddy
beta = 2e-11
f = [2 * (2*np.pi/86400) * np.sin(np.pi/180 * x) for x in lat]

rad = [np.median(x) for x in Ls] # eddy radius

cx = cy = np.ma.masked_all_like(lat) # eddy center locations
lam = np.ma.masked_all_like(lat) # lambda = eddy radius / deformation radius
Ld = np.zeros_like(lat) # deformation radius
Ro = np.ma.masked_all_like(lat) # rossby number
Rh = np.ma.masked_all_like(lat) # rhines number
waterdepth = np.zeros_like(lat) # water depth at eddy center
ind = [] # list of indices

# list comprehension for calculation?
#Ro = [np.median(x/y/z) for x in U[indices] for y in f[indices] for z in L[indices]]# filter out eddies here

# first pass: take out everything north of lat_min
lat_min = 35
indices = np.array([x.mean()>lat_min for x in lat.tolist()])

# ugh... I need a better way of doing this
for ii in np.arange(0,len(lat))[indices]:
    # depths are negative, so use max to get shallowest
    ix = [ find_nearest(topo.x, mm) for mm in lon[ii].ravel() ]
    iy = [ find_nearest(topo.y, mm) for mm in lat[ii].ravel() ]
    waterdepth[ii] = np.max(topo.z[ix,iy])
    
    # get deformation radius
    ix = [ find_nearest(rossx[0,:], mm+360) for mm in lon[ii].ravel() ]
    iy = [ find_nearest(rossy[:,0], mm) for mm in lat[ii].ravel() ]
    Ld[ii] = np.median(defrad[ix,iy])
    
    # filter out based on water depth
    if waterdepth[ii] < -3000 or waterdepth[ii] > -200: continue
    
    ind.append(ii)
    cx[ii] = np.mean(lon[ii])
    cy[ii] = np.mean(lat[ii])
    Ro[ii] = np.median( U[ii])/np.median(f[ii])/rad[ii]
    Rh[ii] = np.median( U[ii] )/beta/rad[ii]**2
    lam[ii] = rad[ii]/Ld[ii]

Ro = Ro.compressed()
Rh = Rh.compressed()
lam = lam.compressed()
```

```python
# plot histograms
Rorange = np.arange(0,0.15,0.04)

plt.figure()
plt.hold(True)
plt.contour(topo.x, topo.y, topo.z.T, levels=[0,-100, -500, -1000, -2000, -4000], colors='k', alpha=0.9)
plt.ylim((30,50))
plt.xlim((-80,-50))
plt.axhline(lat_min, color='k', alpha=0.5)
plt.title('Map showing filtered eddy trajectories')
for ii in ind:
    plt.plot(lon[ii].T, lat[ii].T, alpha=0.6)

plt.figure()
plt.subplot(1,2,1)
plt.hist(Ro, color=colors[2])
plt.xlabel('Rossby number')
plt.ylabel('Number of eddies')
plt.xticks(Rorange)

plt.subplot(1,2,2)
plt.hist(Rh, bins=np.arange(0,Rh.max().round()+1,1), color=colors[2])
plt.xlabel('Rhines number')
plt.axvline(1)
plt.tight_layout()

plt.figure()
plt.subplot(1,2,1)
plt.hist(lam, color=colors[2])
plt.ylabel('Number of eddies')
plt.xlabel('Eddy radius / $L_D$')

plt.subplot(1,2,2)
plt.plot(Ro, Rh,'.')
plt.ylabel('Rh')
plt.xlabel('Ro')
plt.xticks(Rorange)
plt.tight_layout()
```

```python
latmean = [x.mean() for x in lat.tolist()]
plt.plot(np.array(Ld)/1000, np.array(rad)/1000, '.')
plt.xlabel('$L_D$ (km)')
plt.ylabel('Radius (km)')
plt.xlim((10,40))
```

The Rossby number plot looks suspicious at first because the values are pretty low but this is in agreement with the published plots in Chelton et al. (2011). If, based on Olson (1991) review paper, we assume that the Rossby numbers should be closer to O(0.1), this might indicate that either the length scale is overestimated or that the velocity scale is underestimated (geostrophic approximation). Thus, the Rhines number could be underestimated too.

Here, I use $L_s$ as the length scale. From Chelton et al. (2011), 
> The speed-based definition of the eddy scale $L_s$ is the radius of the circle that has the same area as the region within the closed contour of SSH with maximum average speed. ...
> 
> Only about 1% of the extratropical eddies and about 10% of the tropical eddies had Ro > 0.1.
