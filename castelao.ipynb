# Non-dimensional parameters for eddies

Look at Renato's dataset. My goal is to do some non-dimensional parameter statistics

1. Compare eddy radius $L$ to deformation radius $L_R$ in region
2. Rossby, Rhines number statistics
3. Radius / Slope width


```python
%matplotlib inline
%load_ext secnum
%secnum

import numpy as np
import scipy as sp
import h5py
import netCDF4 

import numexpr as ne
from scipy.io import loadmat
import scipy.interpolate

from datetime import datetime
import time
from IPython import display
import pylab

# matplotlib section
import matplotlib as mpl
import matplotlib.pyplot as plt
# big figures
mpl.rcParams['savefig.dpi'] = 2 * mpl.rcParams['savefig.dpi']
# better color cycling
import brewer2mpl
colors =  brewer2mpl.get_map('Set2', 'qualitative', 8).mpl_colors
mpl.rcParams['axes.color_cycle'] = colors
```

```python
# load Castelao datasets

# this file has trajectories
#file2 = 'data/castelao/03eddy_trajectories_7days_gap40km_1992_Leff_mod_MAB_v5.mat'
#trajs = loadmat(file2)
# trajs['eaL'] is Leff in Chelton et al. (2011) notation. 
# Convert to L - length scale for rel vor = 0 contour / radius of max. velocity using formula in that paper : 
#              L = 0.44 * Leff
# also convert to metres
#L = trajs['eaL'][0]*1.64*1000
#U = trajs['eaV'][0]
#lon = trajs['eaX'][0]
#lat = trajs['eaY'][0]

# not sure how this file is different from the trajectory file - seems to be a superset
filename = 'data/castelao/aviso_data_1992_2011_eddies_anticyclones_7days_40km_gap_Leff_mod_MAB.mat'
eddies = loadmat(filename)
U = eddies['aMAXVEL'][0].T
# Quoting Chelton et al. (2011)
#     The speed-based definition of the eddy scale Ls is the radius
#     of the circle that has the same area as the region within the closed
#     contour of SSH with maximum average speed. 
L = eddies['aLs'][0].T * 1000
lat = eddies['aYYcent'][0].T
lon = eddies['aXXcent'][0].T

# load Chelton et al. (1998) Rossby radius atlas
rossfile = '../datasets/rossbyradiusatlas.mat'
ross = loadmat(rossfile)
# grid form for griddata call later
rossgrid = np.c_[ross['lon'], ross['lat']]

# ugh... I need a better way of doing this
# estimate Ro, Rh, deformation radius and reduce to one value per eddy
beta = 2e-11
f = np.zeros_like(lat) # estimated in loop
rad = np.zeros_like(lat) # eddy radius
Lr = np.zeros_like(lat) # deformation radius
Ro = np.zeros_like(lat) # rossby number
Rh = np.zeros_like(lat) # rhines number

for ii in range(len(lat)):
    f[ii] = 2 * (2*np.pi/86400) * np.sin(np.pi/180 * lat[ii])
    rad[ii] = np.median(L[ii])/2 # L is an estimate of diameter
    Lr[ii] = np.median(sp.interpolate.griddata(rossgrid, ross['L'], (360+lon[ii], lat[ii]), method='nearest'))
    Ro[ii] = np.median( U[ii]/f[ii]/L[ii] )
    Rh[ii] = np.median( U[ii]/beta/L[ii]**2 )
```

```python
# plot histograms
plt.figure()
plt.subplot(1,2,1)
plt.hist(Ro, color=colors[2])
plt.xlabel('Rossby number')
plt.subplot(1,2,2)
plt.hist(Rh, color=colors[2])
plt.xlabel('Rhines number')
plt.axvline(1)
plt.tight_layout()

plt.figure()
plt.subplot(1,2,1)
plt.hist(rad/Lr, color=colors[2])
plt.xlabel('Eddy radius / L_R')
plt.subplot(1,2,2)
plt.plot(Ro, Rh,'.')
plt.ylabel('Rh')
plt.xlabel('Ro')
plt.tight_layout()
```

The Rossby number plot looks suspicious because the values are pretty low but this is in agreement with the published plots in Chelton et al. (2011). If based on Olson (1990), we assume that the Rossby numbers should be closer to 0.25, this might indicate that either the length scale is overestimated or the velocity scale is underestimated. Thus, the Rhines number could be underestimated too. For length scale, I use $L_s$. From the paper, 
> The speed-based definition of the eddy scale $L_s$ is the radius of the circle that has the same area as the region within the closed contour of SSH with maximum average speed. ...
> 
> Only about 1% of the extratropical eddies and about 10% of the tropical eddies had Ro > 0.1.

Radius estimate seems to agree pretty well with the deformation radius $L_R$ atlas of Chelton et al. (1998). So, comparing $L_R$ to slope width is a good proxy for eddy radius to slope width ratio.
